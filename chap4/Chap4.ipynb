## 30. 形態素解析結果の読み込み

````python
with open('../data/neko.txt.mecab', encoding='utf-8') as f:
    l = f.read().split('EOS\n')
f.close()
````

````python
ls = list(filter(lambda x: x != '',l))
txtArr=[]
#surface \t part-of-speech, sub-class1, sub-class2, sub-class3, inflection, conjugation, base, reading, pronunciation
for line in ls:
    arr=[]
    for w in line.split('\n'):
        if w == '':continue
        surf,feat = w.split('\t')
        feat=feat.split(',')
        arr.append(
            dict({
                    'surface': surf,
                    'base'   : feat[6],
                    'pos'    : feat[0],
                    'pos1'   : feat[1]
            })
        )
    txtArr.append(arr)
````

## 31. 動詞

````python
resV=[]
for line in txtArr:
    vl=list(filter(lambda x: x['pos'] == '動詞',line))
    resV+=[v['surface'] for v in vl]

print(list(set(resV)))
````

## 32. 動詞の原形

````python
resBase=[]
for line in txtArr:
    vl=list(filter(lambda x: x['pos'] == '動詞',line))
    resBase+=[v['base'] for v in vl]

print(list(set(resBase)))
````

## 33. 「AのB」

````python
resNo=[]
for line in txtArr:
    if len(line)<3:continue
    for i in (1,len(line)-2):
        if line[i-1]['pos']=='名詞' and (line[i]['pos']=='助詞' and line[i]['surface']=='の') and line[i+1]['pos']=='名詞':
            prevNoun,latterNoun=line[i-1]['surface'],line[i+1]['surface']
            resNo.append(f'{prevNoun}の{latterNoun}')

print(list(set(resNo)))
````

## 34. 名詞の連接

````python
import re
resNounRow=[]
for line in txtArr:
    res=[]
    tmp=[]
    for w in line:
        if w['pos']=='名詞':
            tmp.append(w['surface'])
            if len(tmp) >= 2:
                res.append(''.join(tmp))
        else:
            tmp=[]
        
    resNounRow+=res

maxLen=max([len(x) for x in resNounRow if re.match(r'[^a-zA-Z\']+',x)])
[x for x in resNounRow if len(x)==maxLen]
````

## 35. 単語の出現頻度

````python
import collections

ls1d=[x['base'] for x in sum(txtArr, [])]
collections.Counter(ls1d).most_common()
````

## 36. 頻度上位10語

````python
from matplotlib import rcParams
rcParams['font.family'] = 'Noto Sans CJK JP'

import matplotlib.pyplot as plt
wls=collections.Counter(ls1d).most_common(10)
x=[x for x,y in wls]
y=[y for x,y in wls]
plt.bar(x,y)
````

## 37. 「猫」と共起頻度の高い上位10語

````python
def contains(list, filter):
    for x in list:
        if filter(x):
            return True
    return False
cl=[]
for line in txtArr:
    if contains(line, lambda x: x['base']=='猫'):cl+=[x['base'] for x in line if not x['base']=='猫']
cls=collections.Counter(cl).most_common(10)
x=[x for x,y in cls]
y=[y for x,y in cls]
plt.bar(x,y)
````

## 38. ヒストグラム

````python
awls=collections.Counter(ls1d).most_common()
freq=[int(y) for x,y in awls]

plt.hist(freq, bins=20, range=(1,20))
plt.show
````

## 39. Zipfの法則

````python
import math
x=[math.log(r) for r in range(1,len(freq)+1)]
y=[math.log(f) for f in freq]

plt.scatter(x,y)
plt.show
````
