## 12. 1列目をcol1.txtに，2列目をcol2.txtに保存

## 13. col1.txtとcol2.txtをマージ

```python
dfc1 = pd.read_csv('../data/col1.txt', header=None)
dfc2 = pd.read_csv('../data/col2.txt', header=None)

dfm = pd.concat([dfc1, dfc2], axis=1)
dfm.to_csv('../data/col1_2.txt', sep='\t', index=False, header=None)
```
